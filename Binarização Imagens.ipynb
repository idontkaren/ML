<b>Redução de dimensionalidade utilizando Python</b>

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagem = Image.open('sua_imagem.jpg')
iamgem = imagem.convert('RGB')
pixels = np.array(imagem)
```

```python
imagem_cinza = np.dot(pixels[..., :3], [0.2989, 0.5870, 0.1140]).astype(np.uint8)
```

```python
limiar = 127
imagem_binaria = np.where(imagem_cinza > limiar, 255, 0).astype(np.uint8)
```

```python
# Slavar as imagens reduzidas
Image.fromarray(imagem_cinza).save('imagem_cinza.jpg')
Image.fromarray(imagem_binaria).save('imagem_binaria.jpg')
```

```python
#Exibição das imagens reduzidas
plt.subplot(1, 3, 1)
plt.imshow(pixels)
plt.title('Imagem original')
plt.axis('off')
```

```python
plt.subplot(1, 3, 1)
plt.imshow(imagem_cinza, cmap = 'gray')
plt.title('Imagem escala cinza')
plt.axis('off')
```

```python
plt.subplot(1, 3, 1)
plt.imshow(imagem_binaria, cmap = 'gray')
plt.title('Imagem binária')
plt.axis('off')
```
